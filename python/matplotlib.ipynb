```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
x, y
```

```python
plt.scatter(x, y, c = 'g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph in 2D")
```

```python
plt.plot(x, y, 'g*--')
```

```python
y = x*x
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'b')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'g')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'y')
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
data
```

```python
plt.boxplot(data, vert= True, patch_artist=True)
```

```python
array = [23, 45, 70]
myexplode = [0, 0.1, 0]
colors = ['r', 'g', 'b']
labels = ["Apple", "Banana", "Guava"]
plt.pie(array, colors=colors, labels= labels, labeldistance=0.5, explode=myexplode, startangle=90)
plt.legend()
```
